```python
mapping=open("Proj_UID.csv","r")
projects=[]
for line in mapping:
    #print(cols)
    line=line.strip("\n").split(";")
    projects.append(line[1])
mapping.close()
```

```python
projects=list(set(projects))
```

```python
fpapers=open("Papers-Grid_Min.csv","r",encoding="utf-8")

Papers=[]
ProperPaper=[]
keywords=[]

for line in fpapers:
    if "PRJNA" in line:
        #print(line)
        line=line.strip("\n").split('PRJNA')
        #print(line)
        #paper=line[0].split("SRX")[0].split("SAMN")[0]
        projectID="PRJNA"+line[1].split(",")[0].split(";")[0].split(".pdf")[0].split(")")[0]
        #print(projectID)
        if projectID in projects:
            Papers.append(line[0])
            ProperPaper.append(projectID)
            keywords.append(line[1].split(",")[1::])
fpapers.close()
```

```python
AlreadyAdded=[]
Donepapers=open("MetPapers_Orig.txt","r",encoding="utf-8")

for line in Donepapers:
    line=line.strip("\n")
    AlreadyAdded.append(line)
Donepapers.close()
```

```python
MetPapers=open("MetPapers.txt","w",encoding="utf-8")
added=[]
for i in Papers:
    i=i.replace('"','')
    i=i.strip(",")
    i=i.strip(".")
    i=i.strip("\t")
    #print(i)
    if i=="" or i in added or i in AlreadyAdded:
        continue
    else:
        added.append(i)
        MetPapers.write(i+"\n")
MetPapers.close()
```

#### Extract MeSH Terms

```python
XMLPapers=open("PaperCompendium.txt","r",encoding="utf-8")
titles=[]
MajorFields=[]
MinorFields=[]
while True:
    line = XMLPapers.readline()
    if line is None or line=='':
        break
    else:
        if "<ArticleTitle>" in line:
            title=line.strip("\n").replace("\t","").split(">")[1].split("<")[0]
            title=title.replace('"','')
            title=title.strip(",")
            title=title.strip(".")
            title=title.strip("\t")
            thisMajF=[]
            thisMinF=[]
        elif "<MeshHeadingList>" in line:
            while "</MeshHeadingList>" not in line:
                line = XMLPapers.readline()
                if "<MeshHeading>" in line:
                    MinF=[]
                    while "</MeshHeading>" not in line:
                        line = XMLPapers.readline()
                        if "DescriptorName" in line:
                            thisMajF.append(line.strip("\n").replace("\t","").split(">")[1].split("<")[0])
                        elif "QualifierName" in line:
                            MinF.append(line.strip("\n").replace("\t","").split(">")[1].split("<")[0])
                    thisMinF.append(MinF)
            titles.append(title)
            MajorFields.append(thisMajF)
            MinorFields.append(thisMinF)
    
XMLPapers.close()
```

```python
projectIDs=["PRJNA450123","PRJNA549032","PRJNA548424","PRJNA548313","PRJNA547610","PRJNA545958","PRJNA544938","PRJNA544407","PRJNA543713","PRJNA543861","PRJNA542898",
            "PRJNA542210","PRJNA541440","PRJNA541100","PRJNA541010","PRJNA558209","PRJNA540094","PRJNA561420","PRJNA561420","","PRJNA552945",
            "PRJNA552773","PRJNA552585","PRJNA549968","PRJNA547779","PRJNA540190"]
```

```python
MeSHFile=open("MeSHTerms.csv","w",encoding="utf-8")

for i in range(0,len(titles)):
    if i==19:
        continue
    for j in range(0,len(MajorFields[i])):
            MeSHFile.write(titles[i]+";"+projectIDs[i]+";"+MajorFields[i][j]+";")
            MeSHFile.write(";".join(MinorFields[i][j])+"\n")
MeSHFile.close()
```
