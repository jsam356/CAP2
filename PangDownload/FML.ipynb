```python
import click
import subprocess as sp
from glob import glob
from os.path import basename, isfile
from os import makedirs, remove
from random import shuffle
from urllib.error import HTTPError
from pangea_api import (
    Knex,
    User,
    Organization,
)
```

```python
knex = Knex()
User(knex, "jsa3002@med.cornell.edu", "HDNA,.,.1911Wol").login()
print("Logged")
outdir='.'
org = Organization(knex, 'MegaGenome').get()
grp = org.sample_group('SRA').get()
print("Here")
samples = list(grp.get_samples())
```

```python
for sample in samples:
    click.echo(sample.name, err=True)
    try:
        ar = sample.analysis_result('megagenome::v1::mash').get()
    except:
            continue
    sra = ar.field('10K').get()
    sra_path = f'{outdir}/{sample.name}.msh'
    if isfile(sra_path):
        remove(sra_path)
    sra.download_file(filename=sra_path)
```
